# 02) Bar Chart Visualization

```python
# Load dataset
movie_scores = pd.read_csv('movie_scores.csv')
```

```python
# Create bar plot
pos = np.arange(len(movie_scores['MovieTitle']))
width = 0.3

# plt.bar(x, height, [width])

plt.bar(pos - width / 2, movie_scores['Tomatometer'], width, label='Tomatometer', color ='green')
plt.bar(pos + width / 2, movie_scores['AudienceScore'], width, label='AudienceScore', color ='red')


# Specify ticks
plt.xticks(pos, rotation=10)
plt.yticks(np.arange(0, 101, 20))

# Get current Axes for setting tick labels and horizontal grid
ax = plt.gca()

# Set tick labels
ax.set_xticklabels(movie_scores['MovieTitle'])
ax.set_yticklabels(['0%', '20%', '40%', '60%', '80%', '100%'])

# Add minor ticks for y-axis in the interval of 5
ax.set_yticks(np.arange(0, 100, 5), minor=True)

# Add major horizontal grid with solid lines
ax.yaxis.grid(which='major')
# Add minor horizontal grid with dashed lines
ax.yaxis.grid(which='minor', linestyle='--')

# Add title
plt.title('Movie comparison Bar Chart')

# Add legend
plt.legend()
# Show plot
plt.show()
```

```python
# Import statements
import pandas as sb
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# Load dataset
bills = sns.load_dataset('tips')
```

```python
bills.head()
```

```python
days = ['Thur', 'Fri', 'Sat', 'Sun']
days_range = np.arange(len(days))
smoker = ['Yes', 'No']
bills_by_days = [bills[bills['day'] == day] for day in days]
bills_by_days_smoker = [[bills_by_days[day][bills_by_days[day]['smoker'] == s] for s in smoker] for day in days_range]
```

```python
total_by_days_smoker = [[bills_by_days_smoker[day][s]['total_bill'].sum() for s in range(len(smoker))] for day in days_range]
totals = np.asarray(total_by_days_smoker)
```

```python
# Create figure
plt.figure(figsize=(10, 5), dpi=300)
# Create stacked bar plot
plt.bar(days_range, totals[:, 0], label='Smoker')
plt.bar(days_range, totals[:, 1], bottom=totals[:, 0], label='Non-smoker')
# Add legend
plt.legend()

# Add labels and title
plt.xticks(days_range)
ax = plt.gca()
ax.set_xticklabels(days)
ax.yaxis.grid()
plt.ylabel('Daily total sales in $')
plt.title('Restaurant performance')
# Show plot
plt.show()
```

```python
df = pd.read_csv('titanic_train.csv')
```

```python
df.head()
```

```python
survived = [0, 1]
survived_range = np.arange(len(survived))
Sex = ['male', 'female']

df_by_survived = [df[df['Survived'] == day] for day in survived]
df_by_survived_sex = [[df_by_survived[day][df_by_survived[day]['Sex'] == s] for s in Sex] for day in survived_range]
```

```python
total_by_survived_sex = [[df_by_survived_sex[day][s]['Survived'].count() for s in range(len(Sex))] for day in survived_range]
totals_df = np.asarray(total_by_survived_sex)
```

```python
# Create figure
plt.figure(figsize=(10, 5), dpi=300)
# Create stacked bar plot
plt.bar(survived_range, totals_df[:, 0], label='male')
plt.bar(survived_range, totals_df[:, 1], bottom=totals_df[:, 0], label='female')
# Add legend
plt.legend()

# Add labels and title
plt.xticks(survived_range)
ax = plt.gca()
ax.set_xticklabels(['Not Survived', 'Survived'])
ax.yaxis.grid()
plt.ylabel('Total Rations of Male and Female')
plt.title('Survived/Non Surived Male & Female')
# Show plot
plt.show()
```
